```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np
import math
```

### Loading the Data from csv

```python
df = pd.read_csv("outputmesswerteo43.csv", delimiter=";", decimal= ",")
df['rad'] = df.apply(lambda row: row.Grad*math.pi/180, axis=1)
df
```

### Ploting Data for green light
A polar plot ist made to detect the angle of rotation.

```python
cm = 1/2.54

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'},figsize=(20*cm, 20*cm))

theta = df['rad']
r1 = df['stromstärke_grün_ohne']
r2 = df['stromstärke_grün_mit']

ax.set_theta_zero_location('N')
ax.set_rmax(100)
ax.set_thetamin(0)
ax.set_thetamax(360)
ax.set_rticks(np.arange(0, 100, 20))  # Less radial ticks
ax.set_xticks(df['rad'])
ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)

ax.plot(theta, r1, '-o', color= '#76ff03', label='ohne Küvette')
ax.plot(theta, r2, '-o', color= '#32cb00', label='mit Küvette')

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
          fancybox=True, shadow=True, ncol=5)


plt.show()
```

### Plotting Data for red light

```python
fig, ax2 = plt.subplots(subplot_kw={'projection': 'polar'},figsize=(20*cm, 20*cm))

theta2 = df['rad']
rr1 = df['stromstärke_rot_ohne']
rr2 = df['stromstärke_rot_mit']

ax2.set_theta_zero_location('N')
ax2.set_rmax(100)
ax2.set_thetamin(0)
ax2.set_thetamax(360)
ax2.set_rticks(np.arange(0, 100, 20))  # Less radial ticks
ax2.set_xticks(df['rad'])
ax2.set_theta_zero_location('N')
ax2.set_theta_direction(-1)

ax2.plot(theta2, rr1, '-o', color= '#ff616f', label='ohne Küvette')
ax2.plot(theta2, rr2, '-o', color= '#c4001d', label='mit Küvette')

ax2.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
          fancybox=True, shadow=True, ncol=5)

plt.show()
```
