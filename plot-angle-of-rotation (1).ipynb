```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("outputmesswerteo43.csv", delimiter=";", decimal= ",")
df
```

```python
cm = 1/2.54

fig, (ax1, ax2) = plt.subplots(2, 1,figsize=(45*cm, 20*cm))

x1 = df['Grad']
x2 = df['Grad']
gy1 = df['stromstärke_grün_ohne']
gy2 = df['stromstärke_grün_mit']
ry1 = df['stromstärke_rot_ohne']
ry2 = df['stromstärke_rot_mit']

#X1_Y1_Spline = make_interp_spline(x1, gy1)

#X1_ = np.linspace(x1.min(), x1.max(), 100)
#Y1_ = X1_Y1_Spline(X1_)

ax1.plot(x1, gy1, '-o', color= '#76ff03', markersize ='4', linewidth='1', label = 'Stromstärke bei grünem Licht ohne Küvette')
ax1.plot(x1, gy2, '-o', color='#32cb00', markersize ='4', linewidth='1', label = 'Stromstärke bei grünem Licht mit Küvette')
ax2.plot(x2, ry1, '-o', color= '#ff1744', markersize ='4', linewidth='1', label = 'Stromstärke bei rotem Licht ohne Küvette')
ax2.plot(x2, ry2, '-o', color='#c4001d', markersize ='4', linewidth='1', label = 'Stromstärke bei rotem Licht mit Küvette')

ax1.set(xlabel='Drehwinkel in Grad', ylabel='Stromstärke (µA)')
ax1.grid()
ax1.set_xticks(np.arange(0, 380, 20))
ax1.set_yticks(np.arange(0, 120, 10))
ax1.legend(loc='upper center',fancybox=True, ncol=5)


ax2.set(xlabel='Drehwinkel in Grad', ylabel='Stromstärke (µA)')
ax2.grid()
ax2.set_xticks(np.arange(0, 380, 20))
ax2.set_yticks(np.arange(0, 120, 10))
ax2.legend(loc='upper center',fancybox=True, ncol=5)
plt.show()
```
